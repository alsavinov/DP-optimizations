```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Функция для чтения данных из файла
def read_data(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            data.append(int(line.strip()))
    return data

# Задаем диапазоны значений k
k_values = [100, 800]
n_values = read_data('result_n_k_100.txt')  # Предполагаем, что n одинаковые для всех k

# Считываем данные для каждого k
data_knut = [read_data(f'result_knut_k_{k}.txt') for k in k_values]
data_cht = [read_data(f'result_cht_k_{k}.txt') for k in k_values]
data_ras = [read_data(f'result_ras_k_{k}.txt') for k in k_values]


n = np.array(n_values)
K, N = np.meshgrid(k_values, n)

Z_knut = np.array(data_knut).T
Z_cht = np.array(data_cht).T
Z_ras = np.array(data_ras).T

# график
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Рисуем поверхности для каждого метода
ax.plot_surface(N, K, Z_knut, color='yellow', alpha=1, label='Кнут')
ax.plot_surface(N, K, Z_cht, color='red', alpha=1, label='Convex_hull_trick')
ax.plot_surface(N, K, Z_ras, color='blue', alpha=1, label='Разделяй и властвуй')


ax.set_title('3D график времени выполнения')
ax.set_xlabel('Значение n')
ax.set_ylabel('Значение k')
ax.set_zlabel('Время (ms)')


handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)

plt.show()
```
